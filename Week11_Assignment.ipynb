## Landscape RGB Analysis of Color Variations and Intensities


```python
# type: ignore

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("landscape.jpg")
plt.figure(figsize=(10, 5))
plt.title("Original Image")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(15, 5))
plt.title("Gray Scale")
plt.imshow(gray, cmap="gray")
plt.show()
```

```python
blue, green, red = cv2.split(image)

hist_blue = cv2.calcHist([blue], [0], None, [256], [0, 256])
hist_green = cv2.calcHist([green], [0], None, [256], [0, 256])
hist_red = cv2.calcHist([red], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(15, 5))
plt.title("Blue Channel")
plt.xlabel("Intensity")
plt.ylabel("Frequency")
plt.plot(hist_blue, color="blue")
plt.xlim([0, 256])
```

```python
plt.figure(figsize=(15, 5))
plt.title("Green Channel")
plt.xlabel("Intensity")
plt.ylabel("Frequency")
plt.plot(hist_green, color="green")
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.title("Red Channel")
plt.xlabel("Intensity")
plt.ylabel("Frequency")
plt.plot(hist_red, color="red")
plt.show()
```

```python
mean_b, mean_g, mean_r = np.mean(image, axis=(0, 1))
print(f"Mean:")
print(f"Blue: {mean_b:.2f}")
print(f"Green: {mean_g:.2f}")
print(f"Red: {mean_r:.2f}")
```

```python
median_b, median_g, median_r = np.median(image, axis=(0, 1))
print(f"\nMedian:")
print(f"Blue: {median_b:.2f}")
print(f"Green: {median_g:.2f}")
print(f"Red: {median_r:.2f}")
```

```python
std_b, std_g, std_r = np.std(image, axis=(0, 1))
print(f"\nStandard Deviation:")
print(f"Blue: {std_b:.2f}")
print(f"Green: {std_g:.2f}")
print(f"Red: {std_r:.2f}")
```
